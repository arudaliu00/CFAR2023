```python
def prepare_dataframe(df):

    # Sort the DataFrame to ensure proper lagging and differencing
    df.sort_values(by=['GVKEY', 'fyear'], inplace=True)

    # Calculate 'Lagged_Total_Assets'
    df['Lagged_Total_Assets'] = df.groupby('GVKEY')['Total_Assets'].shift(1)

    # Calculate annual changes in 'Current_Assets' and 'CurrentLiabilities'
    df['Delta_CA'] = df.groupby('GVKEY')['Current_Assets'].diff()
    df['Delta_CL'] = df.groupby('GVKEY')['CurrentLiabilities'].diff()
    df['Stock_return'] =  ( df['Year_End_Closing_Price'] - df.groupby('GVKEY')['Year_End_Closing_Price'].shift(1)) / df.groupby('GVKEY')['Year_End_Closing_Price'].shift(1)


    return df

prepare_dataframe(df)
```

```python
def calculate_financial_variables(df):
    df['OA'] = df['Total_Assets'] - df['Cash_Short_Term_Investment']
    df['OL'] = df['Total_Assets'] - df['Debt_included_in_current_liabilities'] - df['Long_Term_Debt'] - df['Minority_Interests'] - df['Preferred_Stocks'] - df['Common_Equity']
    df['Raw_NOA'] = df['OA'] - df['OL']
    df['NOA'] = df['Raw_NOA'] / df['Lagged_Total_Assets']
    df['Earnings'] = df['Income_Operations'] / df['Lagged_Total_Assets']
    df['Raw_Accruals'] = (df['Delta_CA'] - df['Cash']) - (df['Delta_CL'] - df['Debt_included_in_current_liabilities'] - df['Income_Tax_Payable']) - df['Depreciation_and_Amortization']
    df['Accruals'] = df['Raw_Accruals'] / df['Lagged_Total_Assets']
    df['Cash_Flows'] = df['Earnings'] - df['Accruals']
    df['MV'] = df['Year_End_Closing_Price'] * df['Shares_Outstanding']
    df['BV'] = df['BV_Common_Equity']
    df['Cash'] = df['Cash'] / df['Lagged_Total_Assets']
    df['Equity'] = df['Common_Equity'] / df['Lagged_Total_Assets']
    df['Debt'] = df['NOA'] - (df['Equity'] + df['Cash'])
    df['B/M'] = df['BV'] / df['MV']

    # Beta calculation is omitted due to complexity and data requirements.

    return df
calculate_financial_variables(df)
```

```python
# Here's the combined code from loading the dataset to calculating the row-wise percentages for Panel B:



# Define the SIC code ranges for each industry
industry_ranges = {
    "Agriculture": (1, 99),
    "Mining & Construction": (1000, 1299),
    "Food": (2000, 2111),
    "Textiles and Printing/Publishing": (2200, 2790),
    "Chemicals": (2800, 2824),
    "Pharmaceuticals": (2830, 2836),
    "Extractive": (1300, 1399),
    "Durable Manufacturers": (3000, 3569),
    "Computers": (3570, 3579),
    "Transportation": (4000, 4899),
    "Utilities": (4900, 4999),
    "Retail": (5000, 5999),
    "Financial and other": (6000, 6999),
    "Services": (7000, 7369),
}

# Add additional SIC ranges to the dictionary
additional_sic_ranges = {
    "Mining & Construction": [(1400, 1999)],
    "Chemicals": [(2840, 2899)],
    "Extractive": [(2900, 2999)],
    "Durable Manufacturers": [(3580, 3669), (3670, 3699)],
    "Computers": [(3670, 3679), (7370, 7379)],
    "Financial and other": [(2111, 2199)],
    "Services": [(7380, 9999)],
}

# Function to check if SIC code is within the specified range
def is_in_industry(sic, primary_range, additional_ranges=[]):
    if sic >= primary_range[0] and sic <= primary_range[1]:
        return True
    for r in additional_ranges:
        if sic >= r[0] and sic <= r[1]:
            return True
    return False

# Clean the dataframe by dropping NaN values and filtering out negative NOA values
df_cleaned = ind_df.dropna(subset=['Portfolio'])

industry_portfolio_count = {industry: {portfolio: 0 for portfolio in range(1, 11)} for industry in industry_ranges.keys()}

# Count the number of companies for each Portfolio within each industry group
for index, row in df_cleaned.iterrows():
    for industry, primary_range in industry_ranges.items():
        if is_in_industry(row['sic'], primary_range, additional_sic_ranges.get(industry, [])):
            portfolio = row['Portfolio']
            industry_portfolio_count[industry][portfolio] += 1

# Convert the counts to a DataFrame and calculate the column-wise percentages for Panel A (Distribution of number of companies)
industry_portfolio_count_df = pd.DataFrame(industry_portfolio_count)
panel_a_company_percentages = industry_portfolio_count_df.div(industry_portfolio_count_df.sum()).multiply(100)

# Calculate the row-wise percentages for Panel B (Distribution of number of companies)
panel_b_company_percentages = industry_portfolio_count_df.div(industry_portfolio_count_df.sum(axis=1), axis=0).multiply(100)

# Output the data for both panels (Distribution of number of companies)
panel_a_company_output = panel_a_company_percentages.reset_index().rename(columns={'index': 'Portfolio'})
panel_b_company_output = panel_b_company_percentages.reset_index().rename(columns={'index': 'Portfolio'})

panel_a_company_output, panel_b_company_output
```

```python
# Step 1: Sort firms into size quintiles
df4['Size_Quintile'] = df4.groupby('fyear')['MV'].transform(
    lambda x: pd.qcut(x, 5, labels=False, duplicates='drop')
)

# Step 2: Within each size quintile, sort into book-to-market quintiles (excluding negative book values)
df4 = df4[df4['B/M'] > 0]  # Excluding negative book value firms
df4['BTM_Quintile'] = df4.groupby(['fyear', 'Size_Quintile'])['B/M'].transform(
    lambda x: pd.qcut(x, 5, labels=False, duplicates='drop')
)

# Step 3: Within each of the 25 groups, sort into quintiles based on past 12-month returns
df4['Return_Quintile'] = df4.groupby(['fyear', 'Size_Quintile', 'BTM_Quintile'])['Last_SR'].transform(
    lambda x: pd.qcut(x, 5, labels=False, duplicates='drop')
)
df4['Quintile_Group'] = df4['Size_Quintile'].astype(str) + '_' + df4['BTM_Quintile'].astype(str) + '_' + df4['Return_Quintile'].astype(str)

df4
```

```python
# Group the dataframe by fiscal year and Quintile_Group
grouped = df4.groupby(['fyear', 'Quintile_Group'])

# Calculate equal-weighted benchmark returns
ew_returns = grouped['Stock_return'].mean().reset_index(name='EW_Return_Benchmark')

# Calculate value-weighted benchmark returns
vw_returns = grouped.apply(lambda x: (x['MV'] / x['MV'].sum() * x['Stock_return']).sum()).reset_index(name='VW_Return_Benchmark')

# Merge benchmark returns with the original dataframe
df4 = df4.merge(ew_returns, on=['fyear', 'Quintile_Group'], how='left')
df4 = df4.merge(vw_returns, on=['fyear', 'Quintile_Group'], how='left')

# Calculate the abnormal returns
df4['Abnormal_EW_Return'] = df4['Stock_return'] - df4['EW_Return_Benchmark']
df4['Abnormal_VW_Return'] = df4['Stock_return'] - df4['VW_Return_Benchmark']

columns_to_drop = [col for col in df4.columns if '_x' in col or '_y' in col]
df4.drop(columns=columns_to_drop, inplace=True)


# Display the first few rows of the updated dataframe to confirm the changes
df4.head()
```

```python
# Assuming df4 is the dataframe with all necessary columns

# Group by Decile to calculate Equal-Weighted and Value-Weighted Portfolio Returns
ew_portfolio_returns = df4.groupby('Portfolio')['Stock_return'].mean().reset_index(name='Raw_EW')
vw_portfolio_returns = df4.groupby('Portfolio').apply(lambda x: (x['MV'] / x['MV'].sum() * x['Stock_return']).sum()).reset_index(name='Raw_VW')

# Calculate Equal-Weighted and Value-Weighted Abnormal Portfolio Returns for each Decile
ew_abnormal_portfolio_returns = df4.groupby('Portfolio')['Abnormal_EW_Return'].mean().reset_index(name='Adj_EW')
vw_abnormal_portfolio_returns = df4.groupby('Portfolio').apply(lambda x: (x['MV'] / x['MV'].sum() * x['Abnormal_VW_Return']).sum()).reset_index(name='Adj_VW')

# Calculate T-Statistics for Equal-Weighted and Value-Weighted Abnormal Returns
ew_abnormal_tstats = df4.groupby('Portfolio')['Abnormal_EW_Return'].apply(lambda x: stats.ttest_1samp(x.dropna(), 0)[0]).reset_index(name='EW_Abnormal_T-Stat')
vw_abnormal_tstats = df4.groupby('Portfolio')['Abnormal_VW_Return'].apply(lambda x: stats.ttest_1samp(x.dropna(), 0)[0]).reset_index(name='VW_Abnormal_T-Stat')

# Combine all the results into a single DataFrame
results = ew_portfolio_returns.merge(vw_portfolio_returns, on='Portfolio')
results = results.merge(ew_abnormal_portfolio_returns, on='Portfolio')
results = results.merge(vw_abnormal_portfolio_returns, on='Portfolio')
results = results.merge(ew_abnormal_tstats, on='Portfolio')
results = results.merge(vw_abnormal_tstats, on='Portfolio')

# Display the results similar to the provided table format
results

```

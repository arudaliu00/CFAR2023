```python
df=pd.read_csv('/Users/liuaohan/Desktop/paper3database.csv')
df
```

```python
df=df.drop(columns=['consol','popsrc','indfmt','datafmt','curcdq','costat','GVKEY']) #drop useless data
df
```

```python
# fliter 4th quarter data
df['datafqtr']=df['datafqtr'].str[-1:] #
df1=df[df['datafqtr']=='4']
df1
```

```python
# calculation




df1['MV']=df1['prccq']*df1['cshoq']
df1['BV']=df1['atq']-df1['ltq']

# fliter data with zero asset
a=list(df1[df1['atq']==0]['LPERMNO'])
df1 = df1[~df1['LPERMNO'].isin(a)]

b=list(df1[df1['BV']==0]['LPERMNO'])
df1 = df1[~df1['LPERMNO'].isin(b)]

df1['BM']=df1['BV']/df1['MV']
df1['ROA %']=df1['niq']/df1['atq']*100
df1['ROE %']=df1['niq']/df1['BV']*100
df1['Asset']=df1['atq']
df1['Sales']=df1['saleq']
df1['Net income']=df1['niq']


df1['Sales Growth %']=df1['Sales'].diff()/df1['Sales']*100

df1['SRET1 %']=df1['MV'].diff()/df1['MV']*100
df1['SRET12 %']=df1['MV'].diff(2)/df1['MV']*100
df1['CFROA %']=df1['oancfy']/df1['atq']*100
df1['RDINT %']=df1['rdipq']/df1['atq']*100
df1['CAPINT %']=df1['capxy']/df1['atq']*100
df1['ADINT %']=df1['xsgaq']/df1['atq']*100


df1['G1']=0
df1['G2']=0
df1['G3']=0
df1['G4']=0
df1['G5']=0
df1['G6']=0
df1['G7']=0
df1['G8']=0
df1['GScore']=0


df1
```

```python
# fliter data with BM <=0 or inf
a=list(df1[(df1['BM']<=0)|(df1['BM']>10)]['LPERMNO'])
df2 = df1[~df1['LPERMNO'].isin(a)]


df2
```

```python
# fliter all firms less than 2 years of financial statement
gp = df2.groupby('LPERMNO').aggregate(np.count_nonzero)

tf = gp[gp.datafqtr > 2].reset_index()
Permno=list(tf['LPERMNO'])
df3=df2[df2['LPERMNO'].isin(Permno)]

df3
```

```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv("/content/drive/MyDrive/CFAR_1.csv")
df= df[df['invt'] != 0]
df
```

```python
#Filtering the sic

filtered_df = df[~df['sic'].between(4900, 4999) & ~df['sic'].between(6000, 6999)]
filtered_df['sppe'].fillna(0, inplace=True)
company_counts = filtered_df['GVKEY'].value_counts()
company_over2 = company_counts[company_counts >= 2].index.tolist()
final_df = filtered_df[filtered_df['GVKEY'].isin(company_over2)]

final_df
```

```python
cpi_df = pd.read_csv("/content/drive/MyDrive/CPI.csv")
final_df.rename(columns={'fyear': 'Year'}, inplace=True)
merged_df = final_df.merge(cpi_df, on='Year', how='left')
merged_df
```

```python
rf_df = pd.read_csv("/content/drive/MyDrive/RF.csv")
merged_df = merged_df.merge(rf_df, on='Year', how='left')
```

```python
merged_df['RF'] = merged_df['RF'].str.replace('%', '').astype(float) / 100
merged_df['Real_INVT'] = merged_df['invt'] / merged_df['CPI']
merged_df["Net_Inv"] = merged_df.groupby('GVKEY')['Real_INVT'].diff()
merged_df['HN'] = merged_df['Net_Inv'] / merged_df.groupby('GVKEY')['Real_INVT'].shift(1) #The inventory investment rate

merged_df
```

```python
expanded_df = merged_df

expanded_df['I'] = expanded_df['capx'] - expanded_df['sppe'] # firm-level capital investment
expanded_df['IK'] = expanded_df['I'] / expanded_df.groupby('GVKEY')['ppent'].shift(1) # Physical Capital investment rate
expanded_df['SG'] = (expanded_df.groupby('GVKEY')['sale'].diff() / expanded_df.groupby('GVKEY')['CPI'].shift(1)) / expanded_df.groupby('GVKEY')['sale'].shift(1) #Real sales growth
expanded_df['NS'] = expanded_df['invt'] / expanded_df['sale']
expanded_df['Size'] = expanded_df['prcc_c'] * expanded_df['csho'] # Market Equity
expanded_df['Book_Equity'] = expanded_df['bkvlps'] * expanded_df['csho'] # Book Equity
expanded_df['BM'] = expanded_df['Book_Equity'] / expanded_df['Size']
expanded_df['K_ME'] = expanded_df['ppent'] / expanded_df['Size']  #Physical capital-to-market equity ratio
expanded_df['Book_Liab'] = expanded_df['at'] - expanded_df['Book_Equity']
expanded_df['Val_Firm'] = expanded_df['Size'] + expanded_df['at'] - expanded_df['Book_Equity']
expanded_df['Lev'] = expanded_df['Book_Liab'] / expanded_df['Val_Firm']
expanded_df
```

```python
df_FF3 = pd.read_csv("/content/drive/MyDrive/Monthly_FF3_data.csv")
df_FF3['Year'] = pd.to_datetime(df_FF3['Date'], format='%Y%m').dt.year
yearly_averages = df_FF3.groupby('Year').mean()
yearly_averages.reset_index(inplace=True)
yearly_averages.drop(columns=['Date'], inplace=True)
df2 = pd.merge(expanded_df, yearly_averages, on='Year', how='left')
pd.set_option('display.max_columns', None)
df2.drop(columns=['RF2'], inplace=True)

df2
```

```python
df2_1 = df2[(df2['Year'] >= 1993) & (df2['Year'] <= 2003)]
# Sort the DataFrame by 'Year' and 'HN'
df2_1_sorted = df2_1.sort_values(by=['Year', 'HN'])

# Drop NaN values from 'HN', 'prcc_c', and 'RF' to perform quantile binning
df_sorted = df2_1_sorted.dropna(subset=['HN', 'prcc_c', 'RF'])

# Shift the 'HN' column by one year within each group
df_sorted['Prev_Year_HN'] = df_sorted.groupby('Year')['HN'].shift(1)

# Drop rows where 'Prev_Year_HN' is NaN (first year of each group)
df_sorted = df_sorted.dropna(subset=['Prev_Year_HN'])

# Assign portfolio numbers (1-10) to each row within each year based on 'Prev_Year_HN'
df_sorted['Portfolio'] = df_sorted.groupby('Year')['Prev_Year_HN'] \
    .transform(lambda x: pd.qcut(x, 10, labels=range(1, 11), duplicates='drop'))

# Sort by 'GVKEY' and 'datadate' for return calculation
df_sorted = df_sorted.sort_values(by=['GVKEY', 'Year'])

# Calculate the stock return: (Price_t / Price_t-1) - 1
df_sorted['Previous_prcc_c'] = df_sorted.groupby('GVKEY')['prcc_c'].shift(1)
df_sorted['Stock_Return'] = (df_sorted['prcc_c'] / df_sorted['Previous_prcc_c']) - 1


# Calculate excess stock return: Stock Return - RF
df_sorted['Excess_Stock_Return'] = df_sorted['Stock_Return'] - df_sorted['RF']
df_sorted['Market_Excess_Return'] = df_sorted['Market'] - df_sorted['RF']
df_sorted['Market_Cap'] = df_sorted['prcc_c'] * df_sorted['csho']


df_sorted = df_sorted.dropna(subset=['Previous_prcc_c','Stock_Return'])

# Assuming df is your original DataFrame with necessary data including 'Market_Cap', 'Year', and 'Stock_Return'

# Calculate the total market capitalization for each year
total_market_cap_by_year = df_sorted.groupby('Year')['Market_Cap'].sum()

# Calculate the weight for each stock and its value-weighted return
df_sorted['Weight'] = df_sorted.apply(lambda x: x['Market_Cap'] / total_market_cap_by_year[x['Year']], axis=1)
df_sorted['value_weighted_return'] = df_sorted['Weight'] * df_sorted['Stock_Return']
df_sorted
```

```python
#Table 2 Panel A Equal weighted portfolio

import statsmodels.api as sm
# Assuming df is your DataFrame with portfolios and necessary data

def perform_regression_and_get_stats(sub_df, model_type):
    # Define the regression model based on model_type (CAPM or FF3)
    if model_type == 'CAPM':
        X = sub_df[['Market_Excess_Return']]
    elif model_type == 'FF3':
        X = sub_df[['Market_Excess_Return', 'SMB', 'HML']]
    else:
        raise ValueError("Invalid model type")

    y = sub_df['Excess_Stock_Return']
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()

    # Extracting alpha and its t-statistic
    alpha = model.params['const']
    t_stat = model.tvalues['const']

    return alpha, t_stat

# Initialize dictionaries to store results
excess_returns_stats = {}
capm_alpha_stats = {}
ff3_alpha_stats = {}

# Iterate over each portfolio
for portfolio in df_sorted['Portfolio'].unique():
    portfolio_data = df_sorted[df_sorted['Portfolio'] == portfolio]

    # Excess Return Stats
    excess_return = portfolio_data['Excess_Stock_Return'].mean()
    excess_return_t = portfolio_data['Excess_Stock_Return'].std() / (len(portfolio_data) ** 0.5)
    excess_returns_stats[portfolio] = (excess_return, excess_return_t)

    # CAPM Alpha Stats
    capm_alpha, capm_t = perform_regression_and_get_stats(portfolio_data, 'CAPM')
    capm_alpha_stats[portfolio] = (capm_alpha, capm_t)

    # FF3 Alpha Stats
    ff3_alpha, ff3_t = perform_regression_and_get_stats(portfolio_data, 'FF3')
    ff3_alpha_stats[portfolio] = (ff3_alpha, ff3_t)

# Compile the results into a DataFrame for presentation
results_df = pd.DataFrame({
    'Excess_Return': excess_returns_stats,
    'CAPM_Alpha': capm_alpha_stats,
    'FF3_Alpha': ff3_alpha_stats
})



# Identify the portfolios for Low (1) and High (10) inventory growth
low_portfolio = 1
high_portfolio = 10

# Calculate L-H Spread for Excess Returns, CAPM Alphas, and FF3 Alphas
lh_spread = {
    'Excess_Return': (results_df.loc[low_portfolio, 'Excess_Return'][0] - results_df.loc[high_portfolio, 'Excess_Return'][0],
                      results_df.loc[low_portfolio, 'Excess_Return'][1] - results_df.loc[high_portfolio, 'Excess_Return'][1]),
    'CAPM_Alpha': (results_df.loc[low_portfolio, 'CAPM_Alpha'][0] - results_df.loc[high_portfolio, 'CAPM_Alpha'][0],
                   results_df.loc[low_portfolio, 'CAPM_Alpha'][1] - results_df.loc[high_portfolio, 'CAPM_Alpha'][1]),
    'FF3_Alpha': (results_df.loc[low_portfolio, 'FF3_Alpha'][0] - results_df.loc[high_portfolio, 'FF3_Alpha'][0],
                  results_df.loc[low_portfolio, 'FF3_Alpha'][1] - results_df.loc[high_portfolio, 'FF3_Alpha'][1])
}

# Add L-H Spread to the DataFrame
results_df.loc['L-H'] = lh_spread

# Round all numeric values in the DataFrame to two decimal places
results_df = results_df.applymap(lambda x: (round(x[0], 2), round(x[1], 2)) if isinstance(x, tuple) else x)

lh_row = results_df.loc['L-H']
results_df = results_df.drop('L-H')

# Sort the DataFrame by the index (portfolio number)
results_df = results_df.sort_index()

# Append the 'L-H' row back to the DataFrame
results_df.loc['L-H'] = lh_row

print(results_df)
```

```python
#Table 2 Panel A value_weighted

# Step 1: Aggregate value-weighted returns by year and portfolio
yearly_portfolio_return = df_sorted.groupby(['Year', 'Portfolio','Market_Excess_Return','SMB', 'HML']).apply(
    lambda x: np.sum(x['Market_Cap'] / x['Market_Cap'].sum() * x['Stock_Return'])
).reset_index(name='Aggregated_Return')

# Step 2: Calculate annual excess return
yearly_portfolio_return = yearly_portfolio_return.merge(df_sorted[['Year', 'RF']].drop_duplicates(), on='Year')


yearly_portfolio_return['Excess_Return'] = yearly_portfolio_return['Aggregated_Return'] - yearly_portfolio_return['RF']

# Initialize dictionaries for storing results
average_excess_returns = {}
capm_alphas = {}
ff3_alphas = {}

# Step 3 and 4: Iterate over each portfolio
for portfolio in df_sorted['Portfolio'].unique():
    portfolio_data = yearly_portfolio_return[yearly_portfolio_return['Portfolio'] == portfolio]

    # Average excess return and t-statistic
    avg_excess_return = portfolio_data['Excess_Return'].mean()
    t_stat = avg_excess_return / (portfolio_data['Excess_Return'].std() / np.sqrt(len(portfolio_data)))
    average_excess_returns[portfolio] = (avg_excess_return, t_stat)

    # Function for regression and extracting stats
    def perform_regression_and_get_stats(y, X):
        X = sm.add_constant(X)
        model = sm.OLS(y, X).fit()
        return model.params['const'], model.tvalues['const']

    # CAPM Regression
    capm_alpha, capm_t = perform_regression_and_get_stats(
        portfolio_data['Excess_Return'], portfolio_data['Market_Excess_Return']
    )
    capm_alphas[portfolio] = (capm_alpha, capm_t)

    # FF3 Regression
    ff3_alpha, ff3_t = perform_regression_and_get_stats(
        portfolio_data['Excess_Return'], portfolio_data[['Market_Excess_Return', 'SMB', 'HML']]
    )
    ff3_alphas[portfolio] = (ff3_alpha, ff3_t)

# Step 5: Compile the results
results_df = pd.DataFrame({
    'Average_Excess_Return': average_excess_returns,
    'CAPM_Alpha': capm_alphas,
    'FF3_Alpha': ff3_alphas
})


# Identify the portfolios for Low (1) and High (10) inventory growth
low_portfolio = 1
high_portfolio = 10

# Calculate L-H Spread for Excess Returns, CAPM Alphas, and FF3 Alphas
lh_spread = {
    'Average_Excess_Return': (results_df.loc[low_portfolio, 'Average_Excess_Return'][0] - results_df.loc[high_portfolio, 'Average_Excess_Return'][0],
                      results_df.loc[low_portfolio, 'Average_Excess_Return'][1] - results_df.loc[high_portfolio, 'Average_Excess_Return'][1]),
    'CAPM_Alpha': (results_df.loc[low_portfolio, 'CAPM_Alpha'][0] - results_df.loc[high_portfolio, 'CAPM_Alpha'][0],
                   results_df.loc[low_portfolio, 'CAPM_Alpha'][1] - results_df.loc[high_portfolio, 'CAPM_Alpha'][1]),
    'FF3_Alpha': (results_df.loc[low_portfolio, 'FF3_Alpha'][0] - results_df.loc[high_portfolio, 'FF3_Alpha'][0],
                  results_df.loc[low_portfolio, 'FF3_Alpha'][1] - results_df.loc[high_portfolio, 'FF3_Alpha'][1])
}

# Add L-H Spread to the DataFrame
results_df.loc['L-H'] = lh_spread


# Round results to two decimal places
results_df = results_df.applymap(lambda x: (round(x[0], 2), round(x[1], 2)) if isinstance(x, tuple) else x)

lh_row = results_df.loc['L-H']
results_df = results_df.drop('L-H')

# Sort the DataFrame by the index (portfolio number)
results_df = results_df.sort_index()

# Append the 'L-H' row back to the DataFrame
results_df.loc['L-H'] = lh_row

print(results_df)
```

```python
# Table 3 equal weighted


# Assuming 'grouped_df' is your DataFrame with the two-way sorted groups

def calculate_excess_return_stats(sub_df):
    # Calculate the mean excess return
    excess_return = sub_df['Excess_Stock_Return'].mean()
    # Calculate the t-statistic for excess return
    excess_return_t = excess_return / (sub_df['Excess_Stock_Return'].std() / (len(sub_df) ** 0.5))
    return excess_return, excess_return_t

def perform_regression_and_get_stats(sub_df, model_type):
    # Define the regression model based on model_type (CAPM or FF3)
    if model_type == 'CAPM':
        X = sub_df['Market_Excess_Return']
    elif model_type == 'FF3':
        X = sub_df[['Market_Excess_Return', 'SMB', 'HML']]
    else:
        raise ValueError("Invalid model type")

    y = sub_df['Excess_Stock_Return']
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()

    # Extracting alpha and its t-statistic
    alpha = model.params['const']
    t_stat = model.tvalues['const']

    return alpha, t_stat

# Dictionary to store results
results = {}

# Iterate over each combination of HN and Market Cap groups
for hn_group in ['Low', 'Medium', 'High']:
    for cap_group in ['Small', 'Medium', 'Large']:
        key = (hn_group, cap_group)
        sub_df = grouped_df[(grouped_df['HN_Group'] == hn_group) & (grouped_df['Market_Cap_Group'] == cap_group)]

        if not sub_df.empty:
            # Calculate excess return stats
            excess_return, excess_return_t = calculate_excess_return_stats(sub_df)

            # Calculate CAPM alpha stats
            capm_alpha, capm_alpha_t = perform_regression_and_get_stats(sub_df, 'CAPM')

            # Calculate FF3 alpha stats
            ff3_alpha, ff3_alpha_t = perform_regression_and_get_stats(sub_df, 'FF3')

            # Store results
            results[key] = {
                'Excess_Stock_Return': (excess_return, excess_return_t),
                'CAPM_Alpha': (capm_alpha, capm_alpha_t),
                'FF3_Alpha': (ff3_alpha, ff3_alpha_t)
            }


# Print the final DataFrame
print(results_df)
```

```python
#Table 3 value weighted

# Step 1: Aggregate value-weighted returns
grouped_return = grouped_df.groupby(['Year', 'HN_Group', 'Market_Cap_Group','Market_Excess_Return', 'SMB', 'HML']).apply(
    lambda x: np.sum(x['Market_Cap'] / x['Market_Cap'].sum() * x['Stock_Return'])
).reset_index(name='Aggregated_Return')

# Step 2: Calculate annual excess return
grouped_return = grouped_return.merge(df_sorted[['Year', 'RF']].drop_duplicates(), on='Year')
grouped_return['Excess_Return'] = grouped_return['Aggregated_Return'] - grouped_return['RF']

# Initialize dictionaries for storing results
grouped_stats = {}

def perform_regression_and_get_stats(y, X):
    # Ensure X is a DataFrame and y is a Series
    if isinstance(X, pd.DataFrame) and isinstance(y, pd.Series):
        # Add a constant to X for the regression intercept
        X = sm.add_constant(X)
        # Fit the regression model
        model = sm.OLS(y, X).fit()
        # Extract alpha and its t-statistic
        alpha = model.params['const']
        t_stat = model.tvalues['const']
        return alpha, t_stat
    else:
        raise ValueError("X must be a DataFrame and y must be a Series")


# Step 3 and 4: Iterate over each group
for (hn_group, cap_group), group_data in grouped_return.groupby(['HN_Group', 'Market_Cap_Group']):
    # Average excess return and t-statistic
    avg_excess_return = group_data['Excess_Return'].mean()
    t_stat = avg_excess_return / (group_data['Excess_Return'].std() / np.sqrt(len(group_data)))

    # CAPM and FF3 Regression
    capm_alpha, capm_t = perform_regression_and_get_stats(group_data['Excess_Return'], group_data[['Market_Excess_Return']])
    ff3_alpha, ff3_t = perform_regression_and_get_stats(group_data['Excess_Return'], group_data[['Market_Excess_Return', 'SMB', 'HML']])

    grouped_stats[(hn_group, cap_group)] = {
        'Average_Excess_Return': (avg_excess_return, t_stat),
        'CAPM_Alpha': (capm_alpha, capm_t),
        'FF3_Alpha': (ff3_alpha, ff3_t)
    }

# Step 5: Compile the results
grouped_results_df = pd.DataFrame(grouped_stats).T

# Round results to two decimal places
grouped_results_df = grouped_results_df.applymap(lambda x: (round(x[0], 2), round(x[1], 2)) if isinstance(x, tuple) else x)

grouped_results_df
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Preprocess the data
# One-hot encode the 'GVKEY' to handle the categorical nature of companies
# Create lag features for the previous year's metrics
expanded_df['Last_Year_Lev'] = expanded_df.groupby('GVKEY')['Lev'].shift(1)
expanded_df['Last_Year_Price'] = expanded_df.groupby('GVKEY')['prcc_c'].shift(1)

# Drop rows with NaN values that were created by shifting

df_new = expanded_df.sample(frac=0.1, random_state=0)
df_new.replace([np.inf, -np.inf], np.nan, inplace=True)
df_new.dropna(inplace=True)
print(df_new.shape)

# Define the feature set and the target variable
features = ['Last_Year_Price', 'SG', 'NS', 'IK', 'HN', 'Last_Year_Lev', 'GVKEY']
target = 'prcc_c'

# Separate the features and target variable
X = df_new[features]
y = df_new[target]

# Since 'GVKEY' is categorical, we need to one-hot encode it
column_transformer = ColumnTransformer([
    ('gvkey_ohe', OneHotEncoder(handle_unknown='ignore'), ['GVKEY'])
], remainder='passthrough')

X_transformed = column_transformer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, shuffle=False)

# Initialize and train the Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=0)
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Calculate the Root Mean Squared Error
rmse = mean_squared_error(y_test, y_pred, squared=False)

# Compare the predicted and actual prices
comparison = pd.DataFrame({'Actual_Price': y_test, 'Predicted_Price': y_pred})

# Output the results
rmse, comparison
```
